# Matplotlib

To instal lib, on cmd, run: conda install matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline # To show up the plot on notebook
```

```python
# Variables

x = np.linspace(0,5,11)
y = x**2
```

```python
# Functional

plt.plot(x,y)

# Formating legend and Title

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Creating subplots 

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

Get more flexibility and control on plots

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

<b>Put plot inside a plot

```python
# Create a empty plot
fig = plt.figure()


# Set the position of each plot
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3]) # left, bottom, width, height (range 0 to 1)

#Format each plot
axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

<b>Create subplot next to them

```python
# Set the among of plots
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(5,3)) # figsize = (<weight>,<height>) in inch

# Use index to format each plot
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Plot2')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

# Show up the plots
plt.tight_layout()
```

<b>Create two lines plot

```python
# Create a empty plot
fig= plt.figure()

# Add set of axes to figure
ax = fig.add_axes([0,0,1,1])

# Set each lines data, label name
ax.plot(x, x**2, label = 'X Squared')
ax.plot(x, x**3, label = 'X Cubed')

# Show up the legends
ax.legend()
```

<b>Plot Formatation

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])

# Set each lines data, label name, color (w HEX code or name), line style and marker style
ax.plot(x, x**2, label = 'X Squared', color='#FF8C00', , ls = '--', marker = 'o', ms = 7)

# Set box limits
ax.set_xlim([0,3])
ax.set_ylim([0,10])
```
