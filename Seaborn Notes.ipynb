# <b>Seaborn

To install the lib: On cmd, run: conda install seaborn

```python
import seaborn as sns
%matplotlib inline
```

```python
# Import dataset

tips = sns.load_dataset('tips')

tips.head()
```

# <u>Distribuition Plots

# Histogram

<b>sns.distplot</b>

kde = Show Kernel Density Estimation<br>
bins = among of bars

```python
sns.distplot(tips['total_bill'], kde = False, bins =100)
```

# Joinplot

<b>sns.jointplot</b>

kind = type of visualization
Example:
- (defaut): Dispersion analysis
- hex: Hexagonal distribuition
- kde: Kernel density estimation
- reg: Regression analysis

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

# Pairplot

<b>sns.pairplot</b>

Quick way to visualize the relacionship between variables

!Caution! If the dataset have a lot os variables and row, it could spend a long time to process

hue = Categorize values

```python
sns.pairplot(tips, hue='sex')
```

# <u>Categorical Plots

# Bar Plot

<b>sns.barplot</b>

estimator = function to agregate values --> You can use custom functions

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std)
```

# Count Bar Plot

<b>sns.countplot

```python
sns.countplot(x='sex',data=tips)
```

# Box Plot

<b>sns.boxplot</b>

    Points out of the box are OUTLIERS

Use hue to add another category

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

# Violin Plot

<b>sns.violinplot</b>

It's close to boxplot, but show up info KDE info 

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```

# Strip Plot

<b>sns.stripplot</b>

Show the distribuition of the value in each categories

jitter = expand the distribuition (better to see the density) <br>
hue = add a parameter <br>
dodge = split datas 

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True, hue = 'sex', dodge=True)
```

# Combining Plots

```python
sns.stripplot(x='day',y='total_bill',data=tips, color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

# <U>Matrix Plots

```python
# Import dataset
flights = sns.load_dataset('flights')

flights.head()
```

# HeatMap

<b>sns.heatmap</b>

Can be used with correlation table or among in each category

cmap = color gradient type

```python
tc = tips.corr()

sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year', values='passengers')

sns.heatmap(fp, cmap='coolwarm', linecolor='white',linewidths=1)
```

# ClusterMap

<b>sns.clustermap</b>

Show how each group is close from each other

standard_scaleint
Either 0 (rows) or 1 (columns). Whether or not to standardize that dimension, meaning for each row or column, subtract the minimum and divide each by its maximum.

```python
sns.clustermap(fp,cmap ='coolwarm', standard_scale=1)
```

# <u>Grids

```python
# Import dataset
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
```

# Set type of each PairPlots

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# FacetGrid

<b>g.map(plot type,x, y)</b>

Useful to compare a value's distribuition with a categoric matrix

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# <u>Regression

# Simple line regression

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers= ['o','v'], scatter_kws={'s':100})
```

# Multi line regression

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time', hue='day')
```
